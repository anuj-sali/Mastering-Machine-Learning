```python
## 3: How to do EDA efficiently?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df = pd.read_csv('Iris.csv',index_col='Id')
df.head()
```

```python
# Get target: species distribution
f, ax = plt.subplots(1,2,figsize=(18,8))
df.Species.value_counts().plot.pie(explode=[0,0.1,0.1],autopct='%1.1f%%',ax=ax[0])
ax[0].set_title("Species Distribution")
ax[0].set_ylabel('')

sns.countplot('Species', data=df, ax=ax[1])
ax[1].set_title("Species Bar Plot")

plt.tight_layout()
plt.show()
```
